```python
import pandas as pd
```

```python
leg = pd.read_csv("../data/processed/legacy_set_fixed.csv", low_memory=False)
```

```python
leg = leg.drop("Unnamed: 12", axis=1)
```

```python
rest = pd.read_csv("../data/processed/rest_banda_set.csv", low_memory=False)
```

```python
leg.head()
```

```python
leg_part = leg[["DRUGNAME", "NDA_NUM", "origin"]]
rest_part = rest[["drugname", "nda_num", "prod_ai", "origin"]]
leg_part.columns = ["drugname", "nda_num", "origin"]
```

```python
dfs = [leg_part, rest_part]
banda_set_df = pd.concat(dfs, axis=0, ignore_index=True)
```

```python
banda_set_df.to_csv("../data/processed/banda_set.csv", index=False)
```
