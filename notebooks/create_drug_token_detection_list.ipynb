```python
import pandas as pd
```

```python
# Column information obtained from https://www.nlm.nih.gov/research/umls/rxnorm/docs/techdoc.html#conso
column_names = ["RXCUI", "LAT", "TS", "LUI", "STT", "SUI", "ISPREF", "RXAUI", "SAUI", "SCUI",
                "SDUI", "SAB", "TTY", "CODE", "STR", "SRL", "SUPPRESS", "CVF"]
df = pd.read_csv("../data/dictionaries/RXNCONSO.RRF", sep="|", header=None, names=column_names, index_col=False)
```

### The approximate matching authors from RxNorm wrote in their 2011 paper that their approach consisted of selecting all rows where Ingredient (IN), Precise Ingredient (PIN) or Brand Name (BN) was the TTY (Term type in source). They subsequently created a list from all the STR columns in the row, and used this for the identification of the drug token in the input.

```python
df.head()
```

```python
p_df = df[(df["TTY"] == "BN") | (df["TTY"] == "IN") | (df["TTY"] == "PIN")]
```

```python
drug_detection_list = p_df["STR"].sort_values(ascending=True)
```

```python
drug_detection_list.to_csv("../data/dictionaries/drug_detection_list_peters.csv", header=False, index=False)
```
