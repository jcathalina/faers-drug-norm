```python
import pandas as pd
from random import sample
import string
import re
my_punctuation = string.punctuation.replace("-", "")
```

```python
df = pd.read_csv("../data/processed/validated_drugname_list.csv", usecols=["drugname"])
arr = df.to_numpy().astype('str').flatten()
```

```python
p_arr = [a.lower() for a in arr]
p_arr = [re.sub("[\(\[].*?[\)\]]", "", a) for a in p_arr]
p_arr = [a.translate(str.maketrans(' ', ' ', my_punctuation)) for a in p_arr]
pp_arr = []
for a in p_arr:
    result = ''.join([i for i in a if not i.isdigit()])
    pp_arr.append(result)
# pp_arr = [a.replace(' mg ', '') for a in pp_arr]
# pp_arr = [a.replace(' ml ', '') for a in pp_arr]
# pp_arr = [a.rstrip() for a in pp_arr]
ppp_arr = [" ".join(a.split()) for a in pp_arr]
ppp_arr = [a.rstrip(string.punctuation) for a in ppp_arr]
```

```python
sample(ppp_arr, 10)
```

```python
df = pd.read_csv("../data/processed/validated_drugname_list.csv", index_col=False)
```

```python
df["clean_drugname"] = pd.Series(ppp_arr)
```

```python
df = df.rename(columns={"drugname":"old_drugname", "clean_drugname":"drugname"})
```

```python
df.to_csv("../data/processed/validated_drugname_list_with_cleaned_col.csv", index=False)
```
