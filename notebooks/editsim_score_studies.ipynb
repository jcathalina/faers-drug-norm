```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../results/new_editsim_scores_incl_ord.csv")
```

```python
df = df[~df["rxnorm_concept"].isna()]
```

```python
jacc = df["og_score"]
edit = df["new_score"]
edit_ord = df["new_score_ord"]
```

```python
sns.distplot(jacc, hist=False)
sns.distplot(edit, hist=False)
sns.distplot(edit_ord, hist=False)
plt.legend(["Jaccard's Coeff.", "Edit Dist. Score", "Edit Dist. Score (Ordered)"])
plt.xlabel("Score")
plt.savefig("../results/figures/scoring_diffs_no_bars.png")
plt.show()
```
